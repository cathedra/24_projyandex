# Определение успешности игр

Выявить определяющие успешность игры. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными.

## Изучение данных их файла

```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
df = pd.read_csv('/datasets/games.csv')
df.head()
```

```python
df.info()
```

### Вывод

Из приведенного анализа видно, что в данных у нас 16715 строк и 11 столбцом, разбирая их подробнее:
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB  

Entertainment Software Rating Board - эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

Так же из данных видно, что название столбцов с большой буквы. И в ряде столбцов имеются пропуски: Name, Year_of_Release, Genre, Critic_Score, User_Score, Rating

## Подготовка данных

### Замена названия столбцов

Приведем все столбцы к нижнему регистру

```python
df.columns = df.columns.str.lower()
df.head()
```

### Удаление пропусков и преобразование данных в нужных тип

Столбец year_of_release имеет тип данных float, не смотря на то, что это год, преобразуем его к нужному тип, удалив предварительно пустые строки

```python
df.dropna(subset = ['year_of_release'], inplace = True)
df = df.reset_index(drop=True)
df.head()
```

```python
df['year_of_release'] = df['year_of_release'].astype('int')
df.info()
```

Удалим так же 2 пропущенных значения в столбце genre 

```python
df.dropna(subset = ['genre'], inplace = True)
df = df.reset_index(drop=True)
df.head()
```

```python
df.info()
```

Видим, что теперь, пропусков в основных столбцах нет

Столбец user_score имеет тип object, хотя в нем представлен рейтинг, рассмотрим значения, которые в нем встречаются:

```python
df['user_score'].value_counts().head()
```

Видим, что в рейтингах есть указатель TBD, который скорее всего расшифровывается как "To Be Determined", те рейтинг будет определен. Для таких значений логичнее всего проставить NaN и перевести тип данных в Float, так как с NaN можно делать математические операции по столбцу

```python
df['user_score'] = df.loc[df['user_score'] != 'tbd','user_score']
df['user_score'].value_counts().head()
```

```python
df['user_score'] = df['user_score'].astype('float')
df.info()
```

Рассмотрим значение в столбце rating. Известно, что 
- «EC» («Early childhood») — «Для детей младшего возраста»
- «E» («Everyone») — «Для всех»
- «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»
- «T» («Teen») — «Подросткам»
- «M» («Mature») — «Для взрослых»
- «AO» («Adults Only 18+») — «Только для взрослых»
- «RP» («Rating Pending») — «Рейтинг ожидается»

```python
df['rating'].value_counts()
```

Видим, категорию K-A - это "Kids to Adults" «Для всех от 10 лет и старше», те переименуем это название на E10+

```python
df.loc[df['rating'] == 'K-A','rating'] = 'E10+'
df['rating'].value_counts()
```

### Посчитаем суммарные продажи во всех регионах и запишите их в отдельный столбец

```python
df['sales_total'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
df.head()
```

### Продажи игр по платформам

Рассмотрим. как менялись продажи по платформам

```python
df_platform = df.pivot_table(index=['year_of_release'], columns = 'platform', values = 'sales_total', aggfunc='sum')
df_platform
```

Видим, что большая часть платформ выходила на рынок, развивались, а дальше исчезали. Рассмотрим, за какой характерный срок появляются новые платформы и исчезают старые. 

```python
df_platform.plot(style = 'o-', figsize = (20,7));
```

```python
df_platform2 = df_platform.reset_index()
df_platform2
```

```python
df_platform3 = df_platform.T.reset_index()
df_platform3
```

```python
def plat(p):
    k=0;
    b=0;
    с=0;
    for i in df_platform2.index:
        if b < df_platform2.loc[i,p]:
            b = df_platform2.loc[i,p]
            k += 1
            с = df_platform2.loc[i,'year_of_release']
        else:
            df_platform2.loc[i,p]
    print(p,k,b,с)

df_321 = df_platform3['platform'].apply(plat)
```

Исходя из графиков и анализа, видим, что в среднем пик на которое приходится развитие платформы - около 3-4х лет

```python
print('Среднее время существования платформы: {:.1f}'.format(df_platform.count().median()))
```

Видим, что в средняя продолжительность существования платформы около 7 лет.

### Влияние отзывов и критиков на продажи внутри одной платформы:

Для понимая взаимосвязи возьмем одну из самых популярную платформу за все время X360, и проведем анализ внутри нее:

Отсортируем наши данные за актуальный период по платформе X360

```python
df_AP_X360 = df_AP[df_AP['platform'] == 'X360']
df_AP_X360.head()
```

```python
import matplotlib.pyplot as plt 
```

```python
df_AP_X360.plot(x='critic_score', y='sales_total',kind='scatter', alpha=0.1, title = "Зависимость X360: critic_score, sales_total");

plt.ylim(0, 20)
plt.xlim(0, 100)
```

Сгруппируем значения по медиане и еще раз посмотрим на график:

```python
df_AP_X360.groupby('critic_score')['sales_total'].median().plot(x='critic_score', y='sales_total', title = "Зависимость X360: critic_score, sales_total");

plt.ylim(0, 20)
plt.xlim(0, 100)
```

Из анализа анализа графика прослеживается слабая зависимость между оценкой критиков и продажами игр. Но после рейтинга 80, видно, что чем выше рейтинг, тем выше в средняя продажа игр. Проверим коэфициент корреляции для всех данных и для рейтинга выше 80.

```python
print('Коэффициент корреляции между рейтингом критиков и продажами: {:.2f}'.format(df_AP_X360['critic_score'].corr(df_AP_X360['sales_total'])))
```

```python
print('Коэффициент корреляции между рейтингом критиков выше 80 и продажами: {:.2f}'.format(df_AP_X360['critic_score'][df_AP_X360['critic_score'] > 80].corr(df_AP_X360['sales_total'])))
```

Видим, что коэффициент корреляции для осортированных данных значительно вырос. И все же небольшая взаимосвязь с данных существует. Рост так же связан с маленьким количеством игр, рейтинг у которых больше 80.

## Проверка гипотез

Сформулируем и проверим несколько гипотез: 

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сделаем проверку гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам

H0: средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H1: средние пользовательские рейтинги платформ Xbox One и PC различаются

Сделаем два датафрейма для проверки гипотиз, один с платформой Xbox One другой с PC

```python
pd.set_option('mode.chained_assignment', None)

df_XOne = df_AP[df_AP['platform'] == 'XOne']

df_XOne.dropna(subset = ['user_score'], inplace = True)
df_XOne = df_XOne.reset_index(drop=True)
```

```python
df_XOne
```

```python
df_PC = df_AP[df_AP['platform'] == 'PC']

df_PC.dropna(subset = ['user_score'], inplace = True)
df_PC = df_PC.reset_index(drop=True)
```

Зададим общепринятое пороговое значение 5%

```python
from scipy import stats as st

alpha = 0.05

results = st.ttest_ind(df_XOne['user_score'], df_PC['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Посмотрим на среднее двух выборок:

```python
print('Средний пользовательский рейтинг платформы XOne: {:.2f} '.format(df_XOne['user_score'].mean()))
print('Средний пользовательский рейтинг платформы PC: {:.2f} '.format(df_PC['user_score'].mean()))
```

Полученное значение p-value говорит о том, что хотя cредний пользовательский рейтинг платформы XOne и PC неодинакова, с вероятностью в почти 55% такое или большее различие можно получить случайно. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между средней выручкой

Действительно, видим, что средний пользовательский рейтинги платформ близки.

### Средние пользовательские рейтинги жанров Action и Sports разные.

Сделаем проверку гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам

```python
df_Action = df_AP[df_AP['genre'] == 'Action']

df_Action.dropna(subset = ['user_score'], inplace = True)
df_Action = df_Action.reset_index(drop=True)
```

```python
df_Sports = df_AP[df_AP['genre'] == 'Sports']

df_Sports.dropna(subset = ['user_score'], inplace = True)
df_Sports = df_Sports.reset_index(drop=True)
```

H0: средние пользовательские рейтинги жанров Action и Sports одинаковые

H1: средние пользовательские рейтинги  жанров Action и Sports различаются

```python
results = st.ttest_ind(df_Action['user_score'], df_Sports['user_score'], equal_var = False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Посмотрим на среднее двух выборок:

```python
print('Средний пользовательский рейтинг жанра Экшен: {:.2f} '.format(df_Action['user_score'].mean()))
print('Средний пользовательский рейтинг жанра Спорт: {:.2f} '.format(df_Sports['user_score'].mean()))
```

Из проведенного исследования можно сделать вывод, что средние пользовательские рейтинги жанров Экшен и Спорт различаются, это же и показывает среднее значение.
