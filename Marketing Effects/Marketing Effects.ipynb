#### Retention Rate

Рассчитаем коэфициент удержания пользователей и посмотрим на его динамику во времени

Добавим в таблицу vis данные о первом посещении сайта

```python
first_activity_time = vis.groupby(['uid'])['start_ts'].min()
first_activity_time.name = 'first_activity_time'
vis = vis.merge(first_activity_time, on='uid')
vis.head()
```

Выделим в столбце с датой первого посещения первый месяц

```python
vis['first_month'] = vis['first_activity_time'].astype('datetime64[M]')
vis.head()
```

Найдем время жизни для каждого пользователя сайта

```python
vis['cohort_lifetime'] = (vis['month'] - vis['first_month'])/ np.timedelta64(1, 'M')
vis['cohort_lifetime'] = vis['cohort_lifetime'].round().astype('int') 
vis.head()
```

Сгруппируем данные по когортам, выделяя время жизни и количество уникальных пользователей

```python
cohorts = vis.groupby(['first_month', 'cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
cohorts.head()
```

Выделим для каждой корогорты изначальное количество людей и сгруппируем данные

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_month', 'uid']].rename(columns={'uid':'cohort_users'})
initial_users_count
```

```python
cohorts = cohorts.merge(initial_users_count, on='first_month')
cohorts.head()
```

Рассчитаем retetion и сформируем данные в сводную таблицу, а так же сформируем столбец с типичной когортой усредняя полученные значения

```python
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users']
cohorts.head()
```

```python
retention_pivot = cohorts.pivot_table(index='first_month', columns='cohort_lifetime', values='retention', aggfunc='sum')
retention_pivot.loc['typical_cohort'] = retention_pivot.mean()
retention_pivot .fillna('')
```

Оформим сфодную таблицу в более приятный вид

```python
plt.show()
plt.figure(figsize=(15, 5))
plt.title('Retention Rate')
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray')
plt.xlabel("lifetime");
plt.ylabel("когорты");
```

#### LTV

Посчитаем количество новых покупателей (n_buyers) за каждый месяц:

```python
cohort_sizes = (
    orders.groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_order_month', 'n_buyers']
cohort_sizes
```

Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку.

```python
cohorts = (
    orders.groupby(['first_order_month', 'month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
cohorts.head()
```

Объединим полученные таблицы

```python
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
report.head()
```

Для расчета LTV добавим 2 дополнительных столбца:
   1) Для расчета LTV расчитаем валовую прибыль с маржинальность сервиса — 100%. В данном случае выручка будет равна валовой прибыли
   
   2) Добавим сторбец age - возраст когорты

```python
margin_rate = 1

report['gp'] = report['revenue'] * margin_rate
report['age'] = (report['month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
report.head()
```

Расчитаем LTV для каждой когорты деля валовую прибыль на количество покупателей

```python
report['ltv'] = report['gp'] / report['n_buyers']
report.head()
```

Построим сводную таблицу и оформим ее для лучшего понимания

```python
output = report.pivot_table(index='first_order_month', columns='age', values='ltv', aggfunc='mean').cumsum(axis=1)
output.loc['typical_cohort'] = output.mean()
output.fillna('')
```

```python
plt.figure(figsize=(13, 7))
plt.title('LTV по когортам, у.е.')
y_ticks = output.index[:-1].to_series().dt.strftime('%Y-%m').to_list()+['typical_cohort']
sns.heatmap(output, annot=True, yticklabels = y_ticks, fmt='.2f', linewidths=1, linecolor='gray');
plt.xlabel("жизненный цикл");
plt.ylabel("когорты");
```

Отсортируем по когортам, которые прожили больше 6 месяцев и выведим на экран:

```python
report2017 = report[(report['first_order_month'] < '2018') & (report['age'] < 6)]
```

```python
output6 = report2017.pivot_table(index='first_order_month', columns='age', values='ltv', aggfunc='mean').cumsum(axis=1)

plt.figure(figsize=(15, 5))
plt.title('LTV по когортам, у.е.')
y_ticks = output6.index[:7].to_series().dt.strftime('%Y-%m').to_list()
sns.heatmap(output6, annot=True, yticklabels = y_ticks, fmt='.2f', linewidths=1, linecolor='gray');
plt.xlabel("жизненный цикл");
plt.ylabel("когорты");
```

```python
output6.T
```

**Вывод:**
**Из данных видно, что самая прибыльная когорта эта сентябрьская, она достигла самых высоких показателей за 6 месяцев, а уже октябрьская когорта оказалась самой слабой, имея прибыль на клиента в два раза меньше, чем октябрьская. Так же видно, что июньская когорта, растет быстрее всего, имея самый низкий стар на 6 месяц жизни, она достигает наибольших результатов (исключая сентябрьскую когорту). Так же из анализа Retention Rate видно, что с каждой новой когортой поведение покупателей не сильно меняется. Так же стоит обратить внимание, что для каждой когорты с течением жизни коэффициент удержания падает.
